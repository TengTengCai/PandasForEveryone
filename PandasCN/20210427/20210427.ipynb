```python
df = pd.read_csv('zte_cscfports.csv', index_col=0)
df['SIG.RecvKB.low'] = df['SIG.RecvKB'] * 0.6
df['SIG.RecvKB.high'] = df['SIG.RecvKB'] * 1.4
df['SIG.SendKB.low'] = df['SIG.SendKB'] * 0.6
df['SIG.SendKB.high'] = df['SIG.SendKB'] * 1.4
```

```python
DATETIME_FORMAT = '%Y%m%d%H%M%S'
def datetime_to_timestage(data_time):
    try:
        dt = datetime.datetime.strptime(str(data_time), DATETIME_FORMAT)
    except Exception as e:
        print(f"strptime {data_time} have error: {e}")
        return None
    dd = dt.date()
    dd_ts = int(time.mktime(dd.timetuple()))
    dt_ts = int(dt.timestamp())
    return (dt_ts - dd_ts) // 900
```

```python
first_datetime = 20210426000000
def get_date_range(first_datetime):
    first_datetime = datetime.datetime.strptime(str(first_datetime), DATETIME_FORMAT)
    start_date = first_datetime.date()
    end_date = start_date + datetime.timedelta(days=1)

    start_date = start_date.strftime(DATETIME_FORMAT)
    end_date = end_date.strftime(DATETIME_FORMAT)
    date_index = pd.date_range(start=start_date, end=end_date, closed='left', freq='15T')
    date_index = date_index.map(lambda x: x.strftime(DATETIME_FORMAT)).to_list()
    df = pd.DataFrame(date_index, columns=['dateTime'], dtype=str)
    return df
```

```python
df4 = pd.DataFrame()
df7 = pd.DataFrame()
for key, value in df.groupby('NEname'):
    first_datetime = value.iloc[0]['dateTime']
    
    datetime_range_df = get_date_range(first_datetime)
    value = value.astype({'dateTime':str})
#     print(pd.concat([datetime_range_df, value], axis=0))
#     print(datetime_range_df)
#     print(value)
    value = pd.merge(datetime_range_df, value, how='left', on='dateTime')
    value['timeStage'] = value['dateTime'].map(datetime_to_timestage)

    
    high_item = value.loc[:, ['timeStage', 'SIG.RecvKB.high']]
    high_item['timeStage'] = high_item['timeStage'].map(lambda x: f'{x}-high')
    high_item = high_item.set_index('timeStage').T
    high_item['NEname'] = key
    high_item.columns.name = ''
    df7 = pd.concat([df7, high_item])
    
    item = value.loc[:, ['timeStage', 'SIG.RecvKB.low']]
    item['timeStage'] = item['timeStage'].map(lambda x: f'{x}-low')
    item = item.set_index('timeStage').T
    item['NEname'] = key
    item.columns.name = ''
    df4 = pd.concat([df4, item])

df4 = df4.set_index('NEname', drop=True)
df7 = df7.set_index('NEname', drop=True)

print(df4, df7)
```
