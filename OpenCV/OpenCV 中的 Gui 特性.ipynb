# 4. 绘图

```python
img = np.zeros((512,512,3), np.uint8)
plt.imshow(img)
plt.show()
```

## 4.1 画线

```python
cv2.line(img, (0, 0), (300, 511), (0, 0, 255), 3)  
# img需要画线的对象，(0, 0)第一个点，(300, 511)第二个点，(0, 255, 0)线条颜色， 1线条粗细单位像素
plt.imshow(img)
plt.show()
```

## 4.2 画矩形

```python
cv2.rectangle(img,(384,0),(510,128),(0,255,0), 3)
# img需要画线点对象，(384, 0)左侧顶点，(510, 128)右侧顶点， (0, 255, 0)线条颜色， 3线条粗细（=-1为填充整个图形）
plt.imshow(img)
```

## 4.3 画圆

```python
cv2.circle(img, (447, 63), 63, (255, 0, 0), 3)
plt.imshow(img)
```

## 4.4 画椭圆

```python
cv2.ellipse(img,(256,256),(100,50), 0, 0, 180, (0,255,0), -1)
# img 对象，椭圆圆心，椭圆水平直径和垂直直径，椭圆旋转角度， 左侧边旋转，右侧边旋转，颜色， 边框粗细与填充
plt.imshow(img)
```

## 4.5 画多边形

```python
pts=np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts=pts.reshape((-1,1,2))
cv2.polylines(img, [pts], 3, 255)
plt.imshow(img)
```

## 4.6 在图片上添加文字

```python
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'OpenCV',(10,500), font, 4,(255,255,255),2)
plt.imshow(img)
```
